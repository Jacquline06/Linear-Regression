```python
import pandas as pd
```

```python
df=pd.read_csv('/content/emp..csv')
df
```

```python
y = df["salary"] #defining independent(x) and dependant(y) variables #2
x1 =df["dept_id"]
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
sns.set()
```

```python
plt.scatter(x1,y) #3
plt.show()
```

```python
import statsmodels.api as sm
```

```python
x = sm.add_constant(x1) #4
x
```

```python
model = sm.OLS(y,x) #5
```

```python
result = model.fit()
```

```python
result.summary() #6
```

```python
yhat = -1.84e+05 + (1916.2273*x1) #8
yhat
```

```python
plt.scatter(x1,y)
plt.plot(x1,yhat,color = "red")
plt.show()
```

```python
yhat1 =-1.84e+05 + (1916.2265*x1)
yhat1
```

```python
plt.scatter(x1,y)
plt.plot(x1,yhat,color = "red")
plt.plot(x1,yhat1,color = "yellow")

plt.show()
```

```python
yhat_org = result.params[0] + (result.params[1]*x1) #9

plt.scatter(x1,y)
plt.plot(x1,yhat,color = "red")
plt.plot(x1,yhat1,color = "yellow")
plt.plot(x1,yhat_org,color = "green")

plt.show()
```
